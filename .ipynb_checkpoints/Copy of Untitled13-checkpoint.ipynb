# 🎓 AI Study System

A modular AI-powered study assistant. Upload documents, chat, generate mind maps, flashcards, lecture notes, and practice interviews.

**Modules:** `pdf_utils`, `embeddings`, `vector_store`, `rag_pipeline`, `mindmap_generator`, `flashcard_generator`, `lecture_notes_generator`, `interview_voice_assistant`

> Works on both **Google Colab** and **local Windows/Mac/Linux**.

```python
# ══════════════════════════════════════════════════════
# CELL 3: Import Modules & Upload Documents
# ══════════════════════════════════════════════════════
import os, sys, json, time, base64, tempfile, re
from IPython.display import display, Audio, Markdown, Javascript
import ipywidgets as widgets

# ── Platform detection ──
IS_COLAB = False
try:
    from google.colab import files as colab_files, output as colab_output
    IS_COLAB = True
    print('📋 Running on Google Colab')
except ImportError:
    colab_files = None
    colab_output = None
    print('📋 Running locally')

# Import our modular packages
from pdf_utils import load_file
from embeddings import get_embeddings
from vector_store import VectorStoreManager
from rag_pipeline import (
    get_llm, ensure_ollama, ask_question,
    extract_topics, group_similar_topics,
    generate_mcqs, parse_text_mcqs, line_by_line_parse
)
from mindmap_generator import generate_mindmap, mindmap_to_text
from flashcard_generator import generate_flashcards, flashcards_to_text
from lecture_notes_generator import (
    extract_audio, transcribe_audio, generate_lecture_notes, process_video_to_notes
)
from interview_voice_assistant import (
    generate_interview_question, evaluate_answer,
    text_to_speech, speech_to_text, audio_to_base64
)

# PDF export imports
try:
    from reportlab.lib.pagesizes import A4
except ImportError:
    subprocess.run([sys.executable, '-m', 'pip', 'install', '-q', 'reportlab'], check=True)
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.units import mm
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, HRFlowable
from reportlab.lib.enums import TA_LEFT, TA_CENTER

from gtts import gTTS
import speech_recognition as sr

# ── Initialize globals ──
llm = get_llm()
store_manager = VectorStoreManager()

print('✅ All modules imported successfully!')

# ── Upload documents (cross-platform) ──
def upload_documents():
    """Upload documents — uses Colab file picker or local file path."""
    if IS_COLAB:
        print('📂 Choose documents (PDF, PPTX, DOCX, TXT, CSV, XLSX)...')
        uploaded = colab_files.upload()
        for filename, content in uploaded.items():
            filepath = f'/content/{filename}'
            with open(filepath, 'wb') as f:
                f.write(content)
            print(f'⏳ Processing {filename}...')
            docs = load_file(filepath, filename)
            if not docs:
                print(f'❌ No text in {filename}')
                continue
            result = store_manager.add_documents(docs, filename)
            print(f"✅ {filename} — {result['pages']} pages, {result['chunks']} chunks")
    else:
        # Local mode: enter file paths manually
        print('📂 Enter file paths (one per line, empty line to finish):')
        print('   Supported: PDF, PPTX, DOCX, TXT, CSV, XLSX')
        paths = []
        # Try tkinter file dialog first
        try:
            import tkinter as tk
            from tkinter import filedialog
            root = tk.Tk(); root.withdraw()
            selected = filedialog.askopenfilenames(
                title='Select Study Documents',
                filetypes=[
                    ('All supported', '*.pdf;*.pptx;*.docx;*.txt;*.csv;*.xlsx'),
                    ('PDF', '*.pdf'), ('PowerPoint', '*.pptx'),
                    ('Word', '*.docx'), ('Text', '*.txt'),
                    ('CSV', '*.csv'), ('Excel', '*.xlsx'),
                ])
            root.destroy()
            paths = list(selected)
            if paths:
                print(f'📂 Selected {len(paths)} file(s)')
        except Exception:
            pass  # tkinter not available, fall back to manual input

        if not paths:
            # Manual path input fallback
            while True:
                p = input('  Path (or Enter to finish): ').strip().strip('"').strip("'")
                if not p: break
                if os.path.exists(p):
                    paths.append(p)
                else:
                    print(f'  ❌ File not found: {p}')

        for filepath in paths:
            filename = os.path.basename(filepath)
            print(f'⏳ Processing {filename}...')
            docs = load_file(filepath, filename)
            if not docs:
                print(f'❌ No text in {filename}')
                continue
            result = store_manager.add_documents(docs, filename)
            print(f"✅ {filename} — {result['pages']} pages, {result['chunks']} chunks")

    print(f'\n🎉 Loaded: {store_manager.uploaded_docs}')

upload_documents()
```

```python
# ══════════════════════════════════════════════════════
# CELL 4: PDF Export Helpers (Cross-Platform)
# ══════════════════════════════════════════════════════

def download_summaries_as_pdf(summaries):
    """Export summaries to a styled PDF and trigger download."""
    if not summaries: print('⚠️ No summaries!'); return
    out_dir = '/content' if IS_COLAB else os.getcwd()
    path = os.path.join(out_dir, 'summary.pdf')
    doc = SimpleDocTemplate(path, pagesize=A4,
                            leftMargin=20*mm, rightMargin=20*mm,
                            topMargin=20*mm, bottomMargin=20*mm)
    styles = getSampleStyleSheet()
    title_s = ParagraphStyle('T', parent=styles['Title'], fontSize=18,
                             textColor=colors.HexColor('#1e1b4b'), spaceAfter=6)
    doc_s = ParagraphStyle('D', parent=styles['Heading1'], fontSize=13,
                           textColor=colors.HexColor('#6366f1'),
                           backColor=colors.HexColor('#ede9fe'), spaceAfter=4)
    topic_s = ParagraphStyle('To', parent=styles['Heading2'], fontSize=11,
                             textColor=colors.HexColor('#4f46e5'), spaceAfter=2)
    body_s = ParagraphStyle('B', parent=styles['Normal'], fontSize=9,
                            leading=14, spaceAfter=4, leftIndent=8)
    def clean(t):
        for k,v in {'•':'-','→':'->','✅':'[OK]','❌':'[X]','🔹':'-',
                    '📌':'*','\u2022':'-','\u2019':"'",'\u201c':'"',
                    '\u201d':'"','\u2013':'-','\u2014':'--'}.items():
            t = t.replace(k,v)
        return t.encode('ascii','replace').decode('ascii')
    story = [Paragraph('Study Summary', title_s),
             HRFlowable(width='100%', thickness=1, color=colors.HexColor('#6366f1')),
             Spacer(1, 8*mm)]
    for doc_name, topics in summaries.items():
        story.append(Paragraph(clean(doc_name), doc_s))
        story.append(Spacer(1, 3*mm))
        for topic, summary in topics.items():
            story.append(Paragraph(clean(f'• {topic}'), topic_s))
            for line in summary.split('\n'):
                line = line.strip()
                if not line: story.append(Spacer(1,2*mm)); continue
                try: story.append(Paragraph(clean(line), body_s))
                except: pass
            story.append(Spacer(1, 4*mm))
        story.append(HRFlowable(width='100%', thickness=0.5, color=colors.HexColor('#c7d2fe')))
        story.append(Spacer(1, 6*mm))
    try:
        doc.build(story)
        if IS_COLAB:
            colab_files.download(path)
        else:
            print(f'✅ Summary PDF saved to: {os.path.abspath(path)}')
    except Exception as e: print(f'❌ PDF error: {e}')

def download_mcq_as_pdf(questions):
    """Export MCQ results to a styled PDF and trigger download."""
    if not questions: print('⚠️ No questions!'); return
    out_dir = '/content' if IS_COLAB else os.getcwd()
    path = os.path.join(out_dir, 'mcq_results.pdf')
    doc = SimpleDocTemplate(path, pagesize=A4,
                            leftMargin=20*mm, rightMargin=20*mm,
                            topMargin=20*mm, bottomMargin=20*mm)
    styles = getSampleStyleSheet()
    title_s = ParagraphStyle('T', parent=styles['Title'], fontSize=16,
                             textColor=colors.HexColor('#1e1b4b'), spaceAfter=6)
    q_s = ParagraphStyle('Q', parent=styles['Normal'], fontSize=11,
                         fontName='Helvetica-Bold', spaceAfter=4)
    opt_s = ParagraphStyle('O', parent=styles['Normal'], fontSize=10,
                           leftIndent=16, spaceAfter=2)
    exp_s = ParagraphStyle('E', parent=styles['Normal'], fontSize=9,
                           textColor=colors.HexColor('#16a34a'),
                           leftIndent=16, spaceAfter=6, fontName='Helvetica-Oblique')
    def clean(t):
        for k,v in {'•':'-','→':'->','✅':'[OK]','❌':'[X]',
                    '\u2019':"'",'\u201c':'"','\u201d':'"','\u2013':'-'}.items():
            t = t.replace(k,v)
        return t.encode('ascii','replace').decode('ascii')
    story = [Paragraph('MCQ Quiz Results', title_s),
             HRFlowable(width='100%', thickness=1, color=colors.HexColor('#6366f1')),
             Spacer(1, 6*mm)]
    for i, q in enumerate(questions):
        story.append(Paragraph(clean(f"Q{i+1}. {q['question']}"), q_s))
        for k, v in q['options'].items():
            mark = ' [ANSWER]' if k==q['answer'] else ''
            col_hex = '#16a34a' if k==q['answer'] else '#374151'
            ost = ParagraphStyle(f'O{k}{i}', parent=opt_s, textColor=colors.HexColor(col_hex))
            story.append(Paragraph(clean(f'{k}. {v}{mark}'), ost))
        story.append(Paragraph(clean(f"Explanation: {q.get('explanation','')}"), exp_s))
        story.append(Spacer(1, 3*mm))
    try:
        doc.build(story)
        if IS_COLAB:
            colab_files.download(path)
        else:
            print(f'✅ MCQ PDF saved to: {os.path.abspath(path)}')
    except Exception as e: print(f'❌ PDF error: {e}')

print('✅ PDF export helpers loaded')
```

```python
# ══════════════════════════════════════════════════════
# CELL 5: Notebook TTS Helper (Cross-Platform)
# ══════════════════════════════════════════════════════

def speak_full(text, output_area=None, active=None):
    """Speak text aloud — uses JS audio in Colab, saves file locally."""
    try:
        out_dir = '/content' if IS_COLAB else tempfile.gettempdir()
        audio_path = text_to_speech(text, os.path.join(out_dir, 'answer.mp3'))
        if IS_COLAB:
            b64 = audio_to_base64(audio_path)
            display(Javascript(f"""
            (async () => {{
                const audio = new Audio("data:audio/mp3;base64,{b64}");
                audio.play();
                await new Promise(r => {{ audio.onended = r; }});
            }})();
            """))
            time.sleep(max(2, len(text.split()) * 0.38))
        else:
            # Local: display audio widget
            display(Audio(audio_path, autoplay=True))
            time.sleep(max(2, len(text.split()) * 0.38))
    except Exception as e:
        print(f'⚠️ Speech error: {e}')

print('✅ TTS helper loaded')
```

```python
# ══════════════════════════════════════════════════════
# CELL 6: Interview Tab (Cross-Platform)
# ══════════════════════════════════════════════════════

def make_interview_tab():
    """Create the voice interview practice tab widget."""
    state = {
        'active': False, 'history': [], 'current_q': '',
        'total': 0, 'q_count': 0, 'listening': False, 'processing': False
    }

    out = widgets.Output(layout=widgets.Layout(
        min_height='200px', max_height='320px',
        overflow_y='auto', border='1px solid #ddd', padding='8px'))
    q_display     = widgets.HTML('<i>Press 🎤 Start Interview to begin.</i>')
    feedback_html = widgets.HTML('')
    score_html    = widgets.HTML('<b>Score: 0 / 0</b>')
    status        = widgets.HTML('')
    mic_status    = widgets.HTML('')
    start_btn  = widgets.Button(description='🎤 Start Interview', button_style='success')
    end_btn    = widgets.Button(description='⏹ End Interview', button_style='danger', disabled=True)
    # On local: use text input instead of voice
    answer_input = widgets.Text(placeholder='Type your answer here...',
                                layout=widgets.Layout(width='70%'))
    submit_btn = widgets.Button(description='📤 Submit Answer', button_style='primary', disabled=True)
    skip_btn   = widgets.Button(description='⏭ Skip Question', button_style='warning', disabled=True)

    def set_status(msg, color='black'):
        status.value = f"<i style='color:{color}'>{msg}</i>"
    def set_mic(msg, color='#6366f1'):
        mic_status.value = f"<b style='color:{color};font-size:13px'>{msg}</b>"
    def update_score():
        if state['q_count'] == 0:
            score_html.value = '<b>Score: 0 / 0</b>'; return
        avg = state['total'] / state['q_count']
        color = '#16a34a' if avg >= 7 else '#d97706' if avg >= 4 else '#dc2626'
        score_html.value = f"<b style='color:{color}'>Score: {state['total']} / {state['q_count']*10} (avg {avg:.1f}/10)</b>"

    def speak(text):
        """Speak text if on Colab, otherwise just display it."""
        if IS_COLAB:
            try:
                path = text_to_speech(text, '/content/interview_audio.mp3')
                b64 = audio_to_base64(path)
                display(Javascript(f"""
                (async () => {{
                    const audio = new Audio("data:audio/mp3;base64,{b64}");
                    audio.play();
                    await new Promise(r => {{ audio.onended = r; setTimeout(r, 8000); }});
                }})();
                """))
                time.sleep(max(2, len(text.split()) * 0.40))
            except Exception as e:
                print(f'⚠️ TTS error: {e}')

    def show_question(q):
        state['current_q'] = q
        q_display.value = (
            f"<div style='padding:10px;background:#f5f3ff;"
            f"border-left:4px solid #6366f1;border-radius:6px;font-size:14px'>"
            f"<b>Q{state['q_count']+1}:</b> {q}</div>")
        feedback_html.value = ''
        submit_btn.disabled = False; skip_btn.disabled = False
        answer_input.value = ''
        if IS_COLAB:
            set_mic('🎙️ Type your answer or use voice')
        else:
            set_mic('✏️ Type your answer below and click Submit')
        set_status('✅ Question ready — type your answer!', 'green')
        speak(f'Question {state["q_count"]+1}: {q}')

    def process_answer(answer_text):
        """Process a submitted answer (from text or voice)."""
        if not state['active'] or state['processing'] or not answer_text.strip():
            return
        state['processing'] = True
        set_mic('⏳ Evaluating your answer...', '#d97706')
        try:
            with out: display(Markdown(f'**🧑 You:** {answer_text}'))
            set_status('⏳ Evaluating your answer...', '#d97706')
            score, feedback = evaluate_answer(state['current_q'], answer_text, store_manager, llm)
            state['history'].append((state['current_q'], answer_text, feedback, score))
            state['total'] += score; state['q_count'] += 1
            update_score()
            color = '#16a34a' if score >= 7 else '#d97706' if score >= 4 else '#dc2626'
            emoji = '🌟' if score >= 8 else '👍' if score >= 6 else '💡' if score >= 4 else '📚'
            feedback_html.value = (
                f"<div style='padding:10px;border-radius:6px;"
                f"border-left:4px solid {color};background:#f9f9f9;margin-top:8px'>"
                f"<b>{emoji} Score: {score}/10</b><br>{feedback}</div>")
            with out: display(Markdown(f'**🤖 Feedback ({score}/10):** {feedback}\n\n---'))
            speak(f'Score: {score} out of 10. {feedback}')
            if state['active']:
                set_status('⏳ Generating next question...', '#d97706')
                q = generate_interview_question(store_manager, llm, state['history'])
                show_question(q)
        except Exception as e:
            set_mic(f'⚠️ Error: {e}', '#dc2626')
        state['processing'] = False

    def on_submit(b):
        answer_text = answer_input.value.strip()
        if answer_text:
            process_answer(answer_text)

    # Also allow Enter key to submit
    answer_input.on_submit(lambda w: on_submit(None))

    # Colab voice recording callback
    if IS_COLAB:
        def handle_voice_answer(b64_audio):
            if not state['active'] or state['processing']: return
            set_mic('⏳ Processing voice...', '#d97706')
            try:
                with tempfile.NamedTemporaryFile(suffix='.webm', delete=False) as f:
                    f.write(base64.b64decode(b64_audio)); webm = f.name
                answer = speech_to_text(webm)
                answer_input.value = answer
                process_answer(answer)
            except sr.UnknownValueError:
                set_mic('❓ Could not hear you — try typing instead', '#dc2626')
            except Exception as e:
                set_mic(f'⚠️ Error: {e}', '#dc2626')
        colab_output.register_callback('notebook.interview_voice', handle_voice_answer)

    def on_start(b):
        if not store_manager.is_loaded:
            set_status('⚠️ Upload a document first!', '#dc2626'); return
        state.update({'active': True, 'history': [], 'total': 0,
                      'q_count': 0, 'current_q': '', 'listening': False, 'processing': False})
        start_btn.disabled = True; end_btn.disabled = False
        with out: out.clear_output()
        set_status('⏳ Generating first question...', '#d97706')
        speak('Welcome to your interview. Let us begin.')
        q = generate_interview_question(store_manager, llm, first=True)
        show_question(q)

    def on_skip(b):
        if not state['active']: return
        set_status('⏭ Skipped — generating next...', '#d97706')
        state['q_count'] += 1; update_score()
        q = generate_interview_question(store_manager, llm, state['history'] or None, first=not state['history'])
        show_question(q)

    def on_end(b):
        state['active'] = False
        submit_btn.disabled = skip_btn.disabled = True
        start_btn.disabled = False; end_btn.disabled = True
        total = state['total']; count = state['q_count']
        avg = total / (count * 10) * 100 if count else 0
        emoji = '🏆' if avg >= 80 else '👍' if avg >= 50 else '📚'
        color = '#16a34a' if avg >= 80 else '#d97706' if avg >= 50 else '#dc2626'
        msg = 'Excellent preparation!' if avg >= 80 else 'Good effort!' if avg >= 50 else 'Keep studying!'
        q_display.value = (
            f"<div style='text-align:center;padding:16px;background:#f5f3ff;"
            f"border-radius:10px;border:2px solid #6366f1'>"
            f"<div style='font-size:36px'>{emoji}</div>"
            f"<div style='font-size:20px;font-weight:bold;color:{color}'>"
            f"Final Score: {total}/{count*10} ({avg:.0f}%)</div>"
            f"<div style='color:#6b7280'>{msg}</div></div>")
        feedback_html.value = ''
        set_status('⏹ Interview ended.', '#6b7280')
        speak(f'Interview complete. Score: {total} out of {count*10}. {msg}')

    start_btn.on_click(on_start)
    submit_btn.on_click(on_submit)
    skip_btn.on_click(on_skip)
    end_btn.on_click(on_end)

    return widgets.VBox([
        widgets.HBox([start_btn, end_btn, score_html]),
        q_display,
        widgets.HBox([answer_input, submit_btn, skip_btn]),
        mic_status, feedback_html, status,
        widgets.HTML('<b>📜 History:</b>'), out
    ])

print('✅ Interview tab loaded')
```

```python
# ══════════════════════════════════════════════════════
# CELL 7: Launch Main App (All Tabs — Cross-Platform)
# ══════════════════════════════════════════════════════

def launch_app():
    """Build and display the study app with all tabs."""

    # ── TAB 1: SUMMARY ──
    def make_summary_tab():
        out = widgets.Output()
        topic_ddl = widgets.Dropdown(options=['All files']+store_manager.uploaded_docs,
                                    description='File:', layout=widgets.Layout(width='300px'))
        num_topics = widgets.ToggleButtons(
            options=['Concise (3)','Balanced (5)','Detailed (8)'],
            value='Balanced (5)', description='Topics:')
        detail_lvl = widgets.ToggleButtons(
            options=['Brief','Normal','Paragraph'],
            value='Normal', description='Detail:')
        run_btn = widgets.Button(description='▶ Generate Summary', button_style='primary')
        dl_btn  = widgets.Button(description='📄 Download PDF', button_style='warning', disabled=True)
        status  = widgets.HTML(''); store = [{}]

        def on_run(b):
            run_btn.disabled=True; dl_btn.disabled=True
            status.value='<i>⏳ Summarizing...</i>'; out.clear_output()
            tmap = {'Concise (3)':3,'Balanced (5)':5,'Detailed (8)':8}
            dmap = {'Brief':'2-3 bullets only.','Normal':'4-5 clear bullets.',
                    'Paragraph':'One short paragraph, 3-4 sentences.'}
            max_t = tmap[num_topics.value]; dinst = dmap[detail_lvl.value]
            src = None if topic_ddl.value=='All files' else topic_ddl.value
            docs_list = [src] if src else store_manager.uploaded_docs; all_s = {}
            for doc_name in docs_list:
                try: topics = extract_topics(store_manager)
                except Exception as e:
                    with out: print(f'⚠️ {e}'); continue
                topics = group_similar_topics(topics, max_t); doc_s = {}
                for i,(topic,contents) in enumerate(topics.items()):
                    combined = '\n\n'.join(contents)[:2000]
                    status.value = f'<i>⏳ [{i+1}/{len(topics)}] {topic[:40]}...</i>'
                    try:
                        r = llm.invoke(
                            f'Summarize for a student in {dinst}\nTopic:"{topic[:60]}"\nContent:{combined}')
                        summary = r.content if hasattr(r,'content') else str(r)
                    except Exception as e: summary = f'⚠️ {e}'
                    doc_s[topic] = summary
                    with out: display(Markdown(f'### 📌 {topic}\n{summary}\n---'))
                all_s[doc_name] = doc_s
            store[0]=all_s
            status.value='<i style="color:green">✅ Done!</i>'
            run_btn.disabled=False; dl_btn.disabled=False
        run_btn.on_click(on_run)
        dl_btn.on_click(lambda b: download_summaries_as_pdf(store[0]))
        return widgets.VBox([topic_ddl,num_topics,detail_lvl,
                             widgets.HBox([run_btn,dl_btn]),status,out])

    # ── TAB 2: TEXT CHAT ──
    def make_text_chat_tab():
        history = []
        out = widgets.Output(layout=widgets.Layout(
            min_height='250px',max_height='400px',
            overflow_y='auto',border='1px solid #ddd',padding='8px'))
        t_in = widgets.Text(placeholder='Type question + Enter...',
                            layout=widgets.Layout(width='75%'))
        send = widgets.Button(description='Send ➤', button_style='primary')
        clear = widgets.Button(description='🗑 Clear')
        speak_t = widgets.ToggleButton(value=False, description='🔇 Speak OFF',
                                       layout=widgets.Layout(width='120px'))
        status = widgets.HTML('')
        def on_speak(c):
            speak_t.description = '🔊 Speak ON' if speak_t.value else '🔇 Speak OFF'
            speak_t.button_style = 'success' if speak_t.value else ''
        speak_t.observe(on_speak, names='value')
        def on_send(b):
            q = t_in.value.strip()
            if not q: return
            t_in.value=''; status.value='<i>⏳ Thinking...</i>'
            answer = ask_question(q, store_manager, llm, history)
            with out:
                display(Markdown(f'**🧑 You:** {q}\n\n**🤖 AI:** {answer}\n\n---'))
            if speak_t.value: speak_full(answer, out)
            history.append((q, answer))
            if len(history)>10: history.pop(0)
            status.value='<i style="color:green">✅ Done</i>'
        send.on_click(on_send); t_in.on_submit(on_send)
        clear.on_click(lambda b: (out.clear_output(), history.clear()))
        return widgets.VBox([out,widgets.HBox([t_in,send,clear,speak_t]),status])

    # ── TAB 3: VOICE CHAT (Colab) / TEXT CHAT DUPLICATE (Local) ──
    def make_voice_chat_tab():
        history = []; processing = [False]
        out = widgets.Output(layout=widgets.Layout(
            min_height='250px',max_height='400px',
            overflow_y='auto',border='1px solid #ddd',padding='8px'))
        status = widgets.HTML('')

        if IS_COLAB:
            # Full voice mode on Colab
            active = [False]; fails = [0]; MAX_FAILS = 3
            start_btn = widgets.Button(description='🎤 Start', button_style='success')
            stop_btn = widgets.Button(description='⏹ Stop', button_style='danger', disabled=True)
            def stop_loop():
                active[0]=False; processing[0]=False; fails[0]=0
                start_btn.disabled=False; stop_btn.disabled=True
                status.value='⏹ Stopped.'
            def listen():
                if not active[0]: return
                status.value = '🎤 Listening for 5 seconds...'
                display(Javascript('''
                (async () => {
                    const stream = await navigator.mediaDevices.getUserMedia({audio:true});
                    const rec = new MediaRecorder(stream, {mimeType:"audio/webm"});
                    const chunks = [];
                    rec.ondataavailable = e => chunks.push(e.data);
                    rec.start();
                    await new Promise(r => setTimeout(r, 5000));
                    rec.stop(); stream.getTracks().forEach(t => t.stop());
                    await new Promise(r => rec.onstop = r);
                    const blob = new Blob(chunks, {type:"audio/webm"});
                    const reader = new FileReader();
                    reader.readAsDataURL(blob);
                    await new Promise(r => reader.onload = r);
                    await google.colab.kernel.invokeFunction(
                        "notebook.voice_tab",[reader.result.split(",")[1]],{});
                })();'''))
            def handle_audio(b64):
                if not active[0] or processing[0]: return
                processing[0]=True; ensure_ollama()
                try:
                    with tempfile.NamedTemporaryFile(suffix='.webm',delete=False) as f:
                        f.write(base64.b64decode(b64)); webm=f.name
                    status.value='⏳ Transcribing...'
                    question = speech_to_text(webm)
                    fails[0]=0
                    with out: display(Markdown(f'**🧑 You:** {question}'))
                    status.value='⏳ Thinking...'
                    answer = ask_question(question, store_manager, llm, history)
                    history.append((question,answer))
                    if len(history)>10: history.pop(0)
                    with out: display(Markdown(f'**🤖 AI:** {answer}\n\n---'))
                    status.value='🔊 Speaking...'
                    speak_full(answer)
                except sr.UnknownValueError:
                    fails[0]+=1
                    with out: print(f"❓ Didn't catch that ({fails[0]}/{MAX_FAILS})")
                    if fails[0]>=MAX_FAILS: stop_loop(); return
                except Exception as e:
                    with out: print(f'⚠️ Error: {e}'); stop_loop(); return
                processing[0]=False
                if active[0]: status.value='✅ Done! Listening again...'; time.sleep(1); listen()
            colab_output.register_callback('notebook.voice_tab', handle_audio)
            def on_start(b):
                active[0]=True; fails[0]=0; processing[0]=False
                start_btn.disabled=True; stop_btn.disabled=False
                with out: display(Markdown('---\n🎤 **Voice chat started!**\n---'))
                listen()
            start_btn.on_click(on_start)
            stop_btn.on_click(lambda b: stop_loop())
            controls = widgets.HBox([start_btn, stop_btn])
        else:
            # Local mode: text-based chat (voice not available via browser)
            status.value = '💬 Voice recording requires Colab. Using text mode.'
            t_in = widgets.Text(placeholder='Type question + Enter...',
                                layout=widgets.Layout(width='75%'))
            send = widgets.Button(description='Send ➤', button_style='primary')
            def on_send(b):
                q = t_in.value.strip()
                if not q: return
                t_in.value=''; status.value='<i>⏳ Thinking...</i>'
                answer = ask_question(q, store_manager, llm, history)
                history.append((q,answer))
                if len(history)>10: history.pop(0)
                with out: display(Markdown(f'**🧑 You:** {q}\n\n**🤖 AI:** {answer}\n\n---'))
                speak_full(answer)
                status.value='<i style="color:green">✅ Done</i>'
            send.on_click(on_send); t_in.on_submit(on_send)
            controls = widgets.HBox([t_in, send])

        return widgets.VBox([out, controls, status])

    # ── TAB 4 & 5: MCQ / QUIZ ──
    def make_mcq_tab():
        state = {'questions':[],'current':0,'score':0,'answered':False}
        out = widgets.Output()
        topic_in = widgets.Text(value='all topics',description='Topic:',
                                layout=widgets.Layout(width='300px'))
        num_sl = widgets.IntSlider(value=5,min=2,max=10,description='Qs:',
                                   layout=widgets.Layout(width='300px'))
        diff_tb = widgets.ToggleButtons(options=['easy','medium','hard'],
                                        value='medium',description='Difficulty:')
        gen_btn = widgets.Button(description='🎯 Generate',button_style='primary')
        status = widgets.HTML('')
        progress = widgets.HTML('')
        q_html = widgets.HTML('')
        opt_btns = [widgets.Button(description=f'  {x}',
                                   layout=widgets.Layout(width='100%',height='38px'))
                    for x in ['A','B','C','D']]
        feedback = widgets.HTML('')
        next_btn = widgets.Button(description='Next ➤',button_style='primary',disabled=True)
        score_html = widgets.HTML('')
        export_btn = widgets.Button(description='📄 Export PDF',button_style='warning',disabled=True)
        quiz_box = widgets.VBox([progress,q_html,widgets.VBox(opt_btns),
                                 feedback,widgets.HBox([next_btn,export_btn]),score_html],
                                layout=widgets.Layout(display='none'))
        def show_q():
            q=state['questions'][state['current']]
            tot=len(state['questions']); idx=state['current']
            progress.value=f"<b>Q{idx+1}/{tot}</b> — Score: {state['score']}/{idx}"
            q_html.value=(f"<div style='padding:8px;background:#f5f3ff;"
                          f"border-left:4px solid #6366f1;border-radius:4px'>"
                          f"<b>{idx+1}. {q['question']}</b></div>")
            for btn,k in zip(opt_btns,['A','B','C','D']):
                btn.description=f"  {k}.  {q['options'].get(k,'')}"; btn.button_style=''
                btn.disabled=False; btn.style.button_color=None
            feedback.value=''; next_btn.disabled=True; state['answered']=False
        def make_handler(key):
            def handler(b):
                if state['answered']: return
                state['answered']=True
                q=state['questions'][state['current']]; correct=q['answer']
                for btn,k in zip(opt_btns,['A','B','C','D']):
                    btn.disabled=True
                    btn.style.button_color=('#bbf7d0' if k==correct
                                            else '#fecaca' if k==key else None)
                if key==correct:
                    state['score']+=1
                    feedback.value=f"<div style='color:green;padding:6px'>✅ Correct! {q['explanation']}</div>"
                else:
                    feedback.value=f"<div style='color:red;padding:6px'>❌ Wrong. Answer: {correct}. {q['explanation']}</div>"
                next_btn.disabled=False
                if state['current']==len(state['questions'])-1:
                    next_btn.description='Finish 🏁'
            return handler
        for btn,key in zip(opt_btns,['A','B','C','D']): btn.on_click(make_handler(key))
        def on_next(b):
            state['current']+=1
            if state['current']>=len(state['questions']): show_results()
            else: show_q()
        next_btn.on_click(on_next)
        def show_results():
            s,t=state['score'],len(state['questions']); pct=int(s/t*100)
            e='🏆' if pct>=80 else '👍' if pct>=50 else '📚'
            c='#16a34a' if pct>=80 else '#d97706' if pct>=50 else '#dc2626'
            q_html.value=progress.value=feedback.value=''
            for btn in opt_btns: btn.layout.display='none'
            next_btn.layout.display='none'
            score_html.value=(f"<div style='text-align:center;padding:16px;"
                              f"background:#f5f3ff;border-radius:10px;border:2px solid #6366f1'>"
                              f"<div style='font-size:36px'>{e}</div>"
                              f"<div style='font-size:22px;font-weight:bold;color:{c}'>{s}/{t} ({pct}%)</div>"
                              f"<div style='color:#6b7280'>{'Excellent!' if pct>=80 else 'Good effort!' if pct>=50 else 'Keep studying!'}</div></div>")
            export_btn.disabled=False
            status.value=f"<i style='color:green'>✅ Score: {s}/{t}</i>"
        export_btn.on_click(lambda b: download_mcq_as_pdf(state['questions']))
        def on_generate(b):
            gen_btn.disabled=True; status.value='<i>⏳ Generating MCQs...</i>'
            with out: out.clear_output()
            qs = generate_mcqs(topic_in.value or 'all topics', num_sl.value, diff_tb.value, store_manager, llm)
            if not qs:
                status.value="<i style='color:red'>⚠️ Failed. Try a different topic.</i>"
                gen_btn.disabled=False; return
            state.update({'questions':qs,'current':0,'score':0,'answered':False})
            score_html.value=''
            for btn in opt_btns: btn.layout.display=''
            next_btn.layout.display=''; quiz_box.layout.display=''
            status.value=f"<i style='color:green'>✅ {len(qs)} questions ready!</i>"
            show_q(); gen_btn.disabled=False
        gen_btn.on_click(on_generate)
        return widgets.VBox([
            widgets.VBox([topic_in,num_sl,diff_tb,widgets.HBox([gen_btn]),status]),
            quiz_box, out])

    # ── ASSEMBLE ALL TABS ──
    tab = widgets.Tab()
    tab.children = [
        make_summary_tab(),
        make_text_chat_tab(),
        make_voice_chat_tab(),
        make_mcq_tab(),
        make_mcq_tab(),
        make_interview_tab(),
    ]
    for i,t in enumerate(['📝 Summary','💬 Text Chat','🎙️ Voice Chat',
                           '📋 MCQ','🏆 Quiz','🎤 Interview']):
        tab.set_title(i, t)
    display(widgets.HTML(f"""
        <div style='padding:8px 12px;background:#1e1b4b;color:white;
                    border-radius:8px;margin-bottom:6px'>
            <b>🎓 AI Study System</b>
            <span style='color:#818cf8;margin-left:10px'>📁 {store_manager.uploaded_docs}</span>
        </div>"""))
    display(tab)

launch_app()
```

## 🎬 Lecture Notes from Video
Upload a lecture video to extract audio, transcribe, and generate structured notes.

```python
# ══════════════════════════════════════════════════════
# CELL 10: 🎬 Lecture Notes from Video Demo
# ══════════════════════════════════════════════════════

print('🎬 Select a lecture video file (.mp4, .avi, .mkv, .webm)...')

video_paths = []
if IS_COLAB:
    video_upload = colab_files.upload()
    for filename, content in video_upload.items():
        vpath = f'/content/{filename}'
        with open(vpath, 'wb') as f: f.write(content)
        video_paths.append((vpath, filename))
else:
    try:
        import tkinter as tk
        from tkinter import filedialog
        root = tk.Tk(); root.withdraw()
        selected = filedialog.askopenfilenames(
            title='Select Lecture Videos',
            filetypes=[('Video files', '*.mp4;*.avi;*.mkv;*.webm')])
        root.destroy()
        video_paths = [(p, os.path.basename(p)) for p in selected]
    except Exception:
        p = input('Enter video file path: ').strip().strip('"')
        if p and os.path.exists(p):
            video_paths = [(p, os.path.basename(p))]
        else:
            print('❌ File not found')

for video_path, filename in video_paths:
    print(f'\n📹 Processing: {filename}')
    result = process_video_to_notes(video_path, llm)
    print('\n' + '='*50)
    print('📝 STRUCTURED LECTURE NOTES')
    print('='*50)
    display(Markdown(result['notes']))
    print(f'\n📄 Raw transcript length: {len(result["transcript"])} chars')
```
